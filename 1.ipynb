In this task we have implemented the Linear, Polynomial & Ridge regression on the data provided.


Initially, the Data provided is cleaned so as to check if there are any missing values. The missing values are dropped using the dropna() function.

Linear and Ridge regression are imported sklearn. train_test_split is imported for dividing the given data for training and testing purpose.

linreg is the linear regression model and linreg.coef_ and linreg.intercept_ gives us the coefficient and intercept of the model after the application of the model.

As the value of the coefficient is very close to 1 and the intercept is very close to 0 it can be deduced that x and y values are linearly dependent.

The relationship can also be seen from the graph plot below.

In the plot below, scatter plot is of the given x and y. The red line plot gives us the predicted value of y when we give x as an input to the model.

The score on the training as well as test data is given below.


```python
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Ridge
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
import pandas as pd
from matplotlib import pyplot as plt
import numpy as np

df = pd.read_csv(r"D:\Data Science\Data\task3.csv")
df.columns = ['X', 'Y']
df = df.dropna()

X = df['X'].tolist()
X = np.asarray(X)
y = df['Y'].tolist()
y = np.asarray(y)

X = X.reshape(-1,1)
y = y.reshape(-1,1)

X_train1, X_test, y_train, y_test = train_test_split(X,y, random_state = 0)

linreg = LinearRegression().fit(X_train1, y_train)

print("The Coefficient of the Linear Model is ::", (linreg.coef_[0])[0])
print("The intercept of the Linear Model is ::", linreg.intercept_[0])
print("Score on the training data after performing Linear Regression is ::", linreg.score(X_train1, y_train))
print("Score on the test data after performing Linear Regression is ::", linreg.score(X_test, y_test))

plt.scatter(X_train1, y_train, marker = 'o', alpha = 0.5)
plt.plot(X_train1, (linreg.coef_[0])[0]*X_train1 + linreg.intercept_[0], 'r-')
plt.style.use('fivethirtyeight')
plt.title("Graphical Represenation of Linear Regression model")
plt.rcParams["axes.edgecolor"] = "0.15"
plt.rcParams["axes.linewidth"]  = "1.25"
plt.xlabel("X values")
plt.ylabel("Y values")
plt.plot()
```

Polynomial Regresssion is used when the relationship between the input and the output variable is non-linear. 

Polynomial Regression can be achieved by using the sklearn.preprocessing model. The degree used for the given data is 2.

Initially, the input data is converted to polynomial form by using the PolynomialFeatures.fit_transform(Input Data)

Then linear regression is performed on the transformed data.

The coefficient value in this case is approximately 0 as the relationship between the input and the output data is linear. 

As the coefficient is 0 the degree 2 term is 0 and the graph is again linear as shown below.

The score on the training as well as test data is given below.

```python
poly = PolynomialFeatures(degree = 2)
X_poly = poly.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, random_state = 0)

polyreg = LinearRegression().fit(X_train, y_train)

print("The coefficient of the model is ::", (polyreg.coef_[0])[0])
print("The intercept of the model is ::", polyreg.intercept_[0])
print("Score on the training data after performing Polynomial Regression is ::", polyreg.score(X_train, y_train))
print("Score on the test data after performing Polynomial Regression is ::", polyreg.score(X_test, y_test))

pred = polyreg.predict(poly.fit_transform(X))

plt.scatter(X, y, marker = 'o', alpha = 0.5)
plt.plot(X, pred, 'black')
plt.style.use('fivethirtyeight')
plt.title("Graphical Represenation of Polynomial Regression model")
plt.rcParams["axes.edgecolor"] = "0.15"
plt.rcParams["axes.linewidth"]  = "1.25"
plt.xlabel("X values")
plt.ylabel("Y values")
plt.plot()
```

Ridge Regression is performed on the given data here. The input data is transformed into polynomial form and then ridge regression is performed.

The coefficient value is 0 here as well.

The score on the training as well as test data is given below.

```python
poly = PolynomialFeatures(degree = 2)
X_poly = poly.fit_transform(X)
X_train, X_test, y_train, y_test = train_test_split(X_poly, y, random_state = 0)

ridge = Ridge().fit(X_train, y_train)

print("The coefficient of the model is ::", (ridge.coef_[0])[0])
print("The intercept of the model is ::", ridge.intercept_[0])
print("Score on the training data after performing Ridge Regression is ::", ridge.score(X_train, y_train))
print("Score on the test data after performing Polynomial Regression is ::",ridge.score(X_test, y_test))

pred = ridge.predict(poly.fit_transform(X))

plt.scatter(X, y, alpha = 0.5, marker = 'o')
plt.plot(X, pred, 'purple')
plt.style.use('fivethirtyeight')
plt.title("Graphical Represenation of Ridge Regression model")
plt.rcParams["axes.edgecolor"] = "0.15"
plt.rcParams["axes.linewidth"]  = "1.25"
plt.xlabel("X values")
plt.ylabel("Y values")
plt.plot()
```
